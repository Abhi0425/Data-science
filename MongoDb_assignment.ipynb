MongoDB is a popular NoSQL database management system that stores data in a flexible, JSON-like format known as BSON (Binary JSON). It is classified as a document-oriented database, falling under the category of non-relational databases.

###### Non-relational databases:

Non-relational databases, or NoSQL databases, are a type of database that does not rely on the traditional relational database management system (RDBMS) model.



Scenarios to prefer MongoDB over SQL databases:

Flexible Schema,
scalability,
complex data structure,
Big data and analytics,
real time application,

##### Ans - 2
key features of MongoDB:

###### Document-Oriented:

MongoDB stores data in flexible, JSON-like BSON documents. Each document can have a different structure, and related data can be nested within a document.

###### Schema-less:

MongoDB is schema-less, allowing for dynamic and evolving data models. There's no need to predefine the structure of documents, providing flexibility in handling changing data requirements.

###### Dynamic Schema:

Fields can be added to documents on the fly, and each document in a collection can have a different structure. This dynamic schema is beneficial for accommodating diverse and evolving data.

###### Indexing:

MongoDB supports secondary indexes, improving the query performance by allowing for efficient retrieval of data. Indexes can be created on any field in a document.

##### Ans -6
The sort() method in MongoDB is used to specify the order in which the query results should be returned. It allows you to sort the documents in ascending (1) or descending (-1) order based on one or more fields.Example:



import pymongo


host = "your_mongodb_host"
port = 27017  # default MongoDB port

client = pymongo.MongoClient(host, port)

database_name = "your_database_name"
db = client[database_name]

collection = db[collection_name]

sample_data = [
    {"name": "Alice", "age": 25, "city": "Wonderland"},
    {"name": "Bob", "age": 30, "city": "Example City"},
    {"name": "Charlie", "age": 35, "city": "Testville"}
]

collection.insert_many(sample_data)

query_criteria = {"age": {"$gt": 25}}  # Find documents where age is greater than 25
sort_order = [("age", pymongo.ASCENDING)]  # Sort in ascending order based on age
result_set = collection.find(query_criteria).sort(sort_order)

# Print the sorted query result
print("Sorted Query Result:")
for document in result_set:
    print(document)

# Close the MongoDB connection
client.close()
